```python
# Creating numpy array of all attributes
df = dataset.iloc[:, 1 : -1]
attributes = df.to_numpy()
attributes
```

```python
minAttributes = np.amin(attributes, axis=0)
minAttributes
```

```python
maxAttributes = np.amax(attributes, axis =0)
maxAttributes
```

```python
# Function returns euclidean distance between two vectors
def euclideanDistance(vector1, vector2):
    dist = np.linalg.norm(vector1 - vector2)
    return dist
```

```python
# Function returns the index of the cluster with the shortest distance to the inputted attribute vector
def assignedCluster(attribute):
    clusterIndex = 0
    minDistance = euclideanDistance(attribute, clusterCenters[0])
    for i in range(len(clusterCenters)):
        if (euclideanDistance(attribute, clusterCenters[i]) < minDistance):
            clusterIndex = i
            minDistance = euclideanDistance(attribute, clusterCenters[i])
    return clusterIndex
```

```python
def centroid(attributeArray):
    return np.mean(attributeArray, axis = 0)
```

```python
for i in range (30):
    # Using the min and max arrays as bounds, generate random arrays that fall within these constraints
    clusterCenters = []
    for i in range(numClusters):
        tempCluster = []
        for j in range(minAttributes.size):
            tempCluster.append(random.randint(round(minAttributes[j]), round(maxAttributes[j] + 1)))
        clusterCenters.append(tempCluster)
    clusterCenters

    for i in range(numIterations + 1):
        # Array of index values corresponding to the clusterCenters
        assignedClusters = []
        for attribute in attributes:
            assignedClusters.append(assignedCluster(attribute))


        # Creating dictionary to hold cluster numbers as keys and corresponding attribute indexes as values to group by cluster
        attributesByCluster = {}
        for i in range(numClusters):
            attributeIndexes = []
            for j in range(len(assignedClusters)):
                if (assignedClusters[j] == i):
                    attributeIndexes.append(j)
            attributesByCluster[i] = attributeIndexes



        # Replacing indexes in attributesByCluster by their actual attributes
        for indexList in list(attributesByCluster.values()):
            for i in range(len(indexList)):
                indexList[i] = attributes[indexList[i]]


        # Update clusterCenters
        for i in list(attributesByCluster.keys()):
            if(attributesByCluster[i]):
                clusterCenters[i] = centroid(np.asarray(attributesByCluster[i]))


    actualClusters = dataset.iloc[:,-1].to_numpy()

    # Using NMI as metric to measure model
    from sklearn.metrics.cluster import normalized_mutual_info_score
    print(normalized_mutual_info_score(actualClusters, assignedClusters))
```
